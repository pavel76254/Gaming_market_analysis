Некоторые игры выпускались одновременно на нескольких платформах, поэтому для этих данных, значение даты можно взять с других платформ.

```python
not_null_year = df[df['year_of_release'] != 0]
def fill_year_of_release(row):
    col = 'year_of_release'
    rows = not_null_year[not_null_year['name'] == row['name']]
   
    if len(rows) > 0:
        return rows.iloc[0][col]
    else: return row[col]
```

Заполнение работает поля 'year_of_release' работает довольно долго.

```python
%time df['year_of_release'] = df.apply(fill_year_of_release, axis=1).astype('int')
```

Кол-во выпущенных в зависимости от года.

```python
games_per_year = (
    df[['name', 'year_of_release']]
        .pivot_table(index = 'year_of_release', values = 'name', aggfunc = 'count')
        .sort_values('year_of_release', ascending = False)
)
```

```python
(games_per_year
        .plot(figsize=(12, 6), colormap='jet', legend=False, title='Количество игр по годам')
        .set(xlabel='Год релиза', ylabel='Кол-во')
)

plt.show()
```

```python
# Попробуем посчитать срок жизни игровых платформ по следующему критерию:
# Делаем выборку из данных по платформе,
# Далее считаем кол-во лет, когда на платформе было выпущено более 3-х игр
def platform_lives(df, name, games_count=3):
    platform_games_count = df.query("platform==@name").groupby('year_of_release')['total_sales'].count()
    years_lives = platform_games_count[platform_games_count>games_count].count()
    return years_lives
```

# Шаг 7. Проверьте гипотезы

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные.
Задайте самостоятельно пороговое значение alpha.

Поясните:
* Как вы сформулировали нулевую и альтернативную гипотезы;
* Какой критерий применили для проверки гипотез и почему.


Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Н0 (нулевая гипотеза) - cредние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

Для проверки гипотезы возьмём набор данных actual_period.

```python
xone = actual_period.query('platform == "XOne"')
pc = actual_period.query('platform == "PC"')
```

```python
alpha = .05
```

Возьмём значение стандартное значение alpha 0.05

Проверим равны ли дисперсии сравниваемых выборок.

```python
xone['user_score'].var()
```

```python
pc['user_score'].var()
```

Дисперсии не равны, поэтому для проведения t-критерий Стьюдента, будем использовать параметр equal_var=False.

```python
results = st.ttest_ind(
    xone['user_score'],
    pc['user_score'],
    equal_var=False,  # Welch’s t-test, который не предполагает равенство дисперсий
    nan_policy='omit' # игнорируем пропуски
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Действительно, посчитаем средние значения пользовательского рейтинга для XOne и PC. 

```python
pc['user_score'].mean()
```

```python
xone['user_score'].mean()
```

Средние значения примерно равны, принимаем нулевую гипотезу.

#### Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные.

Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые.

Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются между собой.

```python
alpha = .05
```

Также возьмём стандартное значение параметра alpha=0.05 и проверим равенство дисперсий двух выборок

```python
action = actual_period.query("genre == 'Action'")
sports = actual_period.query("genre == 'Sports'")
```

```python
action['user_score'].var()
```

```python
sports['user_score'].var()
```

Дисперсии выборок опять не равны, поэтому снова будем использовать параметр equal_var = False

```python
results = st.ttest_ind(
    action['user_score'],
    sports['user_score'],
    equal_var=False,
    nan_policy='omit' # игнорируем пропуски
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Проверим средние значения:

```python
action['user_score'].mean()
```

```python
sports['user_score'].mean()
```

Они довольно значительно отличаются.
